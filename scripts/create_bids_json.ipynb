```python
import json
from joblib import Parallel, delayed
from pathlib import Path

from shutil import rmtree
```

```python
path = Path("../../hcp1200")
```

```python
i = 0

d = []

for p in path.iterdir():
    if p.is_dir():
        p = p / "ses-1" / "dwi"
        if not list(p.glob("*")):
            i += 1
            d.append(p)

with open('../removed_subjects.txt', 'w') as f:
    for fname in sorted(d):
        sub = fname.parts[3] + "\n"
        f.write(sub)
```

```python
# Remove folders
for p in path.iterdir():
    if p.is_dir():
        p_ = p / "ses-1" / "dwi"
        if not list(p_.glob("*")):
            rmtree(p)
```

```python
bids_json_dict = dict(
    PhaseEncodingDirection = "i-",
    TotalReadoutTime = 0.11232
)
```

```python
for p in path.iterdir():
    if p.is_dir():
        p = p / "ses-1" / "dwi"
        sub = p.parts[3]

        bids_json_name = sub + "_ses-1_dwi.json"
        bids_json_path = p / bids_json_name

        with open(bids_json_path, "w") as write_file:
            json.dump(bids_json_dict, write_file, indent=4)
    
        print(p)
```
